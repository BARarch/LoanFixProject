```python
def monthsCount(rds):
    return rds.years * 12 + rds.months
```

```python
def monthsBetween(issued, payment):
    if not payment:
        return 0
    return monthsCount(rd.relativedelta(dt.datetime.strptime(payment, '%b-%Y'), dt.datetime.strptime(issued, '%b-%Y')))
```

```python
def termToMonths(term):
    return int(term[1:3])
```
